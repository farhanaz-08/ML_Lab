```python
# -----------------------------
# Step 1: Install and import packages
# -----------------------------
!pip install --upgrade google-generativeai
from IPython.display import Image, display
import google.generativeai as genai

# -----------------------------
# Step 2: Set your Gemini API key
# -----------------------------
genai.configure(api_key="AIzaSyDMKXAGsUwNJDFp3k1FduAKFcaDG4JHYpI")  # replace with your key

# -----------------------------
# Step 3: Upload the image
# -----------------------------
from google.colab import files

uploaded = files.upload()
image_filename = list(uploaded.keys())[0]  # get uploaded file
display(Image(filename=image_filename))

# -----------------------------
# Step 4: Read image bytes
# -----------------------------
with open(image_filename, "rb") as f:
    image_bytes = f.read()

# -----------------------------
# Step 5: Use the chat API to extract text
# -----------------------------
response = genai.chat.create(
    model="gemini-1.5-multimodal-preview",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "input_text", "text": "Extract all text from this image."},
                {"type": "input_image", "image": image_bytes}
            ]
        }
    ]
)

# -----------------------------
# Step 6: Get extracted text
# -----------------------------
extracted_text = response.messages[0].content[0]["text"]
print("ðŸ“„ Extracted Text:\n", extracted_text)
```

```python
# Install and import
!pip install --upgrade google-generativeai

import google.generativeai as genai
from PIL import Image

# Set API key
genai.configure(api_key="AIzaSyDMKXAGsUwNJDFp3k1FduAKFcaDG4JHYpI")

# Load model that supports images
model = genai.GenerativeModel("gemini-pro-vision")

# Upload / load image
from google.colab import files
uploaded = files.upload()
image_filename = list(uploaded.keys())[0]

img = Image.open(image_filename)

# Call generate_content with text + image
response = model.generate_content(
    content=["Extract all text from this image:", img]
)

# Print the result
print("ðŸ“„ Extracted Text:")
print(response.text)
```
