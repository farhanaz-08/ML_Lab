```python
# Root folder of your dataset on Drive
root_dir = "/content/drive/MyDrive/Fake-News-Detection-Bilingual-Multimodel/data"

# Paths for each folder
fake_EN = os.path.join(root_dir, "fake/EN")
fake_BN = os.path.join(root_dir, "fake/BN")
real_EN = os.path.join(root_dir, "real/EN")
real_BN = os.path.join(root_dir, "real/BN")

# Output CSV files
fake_csv = "/content/fake_texts.csv"
real_csv = "/content/real_texts.csv"
```

```python
def extract_text_from_image(img_path, model="gemini-ocr-v1"):
    """
    Given an image path, uses Gemini OCR to extract text.
    """
    try:
        img = Image.open(img_path)  # PIL image
        # Gemini OCR call
        response = genai.ocr(model=model, image=img)  # replace with correct OCR call
        return response['text']
    except Exception as e:
        print(f"‚ùå Error extracting text from {img_path}: {e}")
        return None
```

```python
def process_folder(folder_path, label, output_csv, model="gemini-ocr-v1"):
    """
    Loop through all images in folder_path, extract text using Gemini, save to CSV with label.
    """
    files = [f for f in os.listdir(folder_path) if f.lower().endswith((".png", ".jpg", ".jpeg"))]
    print(f"üì∏ Found {len(files)} images in {folder_path}")

    with open(output_csv, "a", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        for filename in files:
            img_path = os.path.join(folder_path, filename)
            print(f"Processing: {img_path}")
            text = extract_text_from_image(img_path, model=model)
            if text:
                writer.writerow([text, label])
```

```python
# Clear CSVs first (optional)
open(fake_csv, "w").close()
open(real_csv, "w").close()

# Process Fake images
process_folder(fake_EN, "fake", fake_csv)
process_folder(fake_BN, "fake", fake_csv)

# Process Real images
process_folder(real_EN, "real", real_csv)
process_folder(real_BN, "real", real_csv)
```
