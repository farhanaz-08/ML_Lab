Milestone 2: Data Preparation & Preprocessing
- Handle any missing values, duplicates, or outliers in the dataset
- Apply feature scaling/normalization
- Encode categorical variables (if present) with a suitable technique
- Perform basic correlation analysis and point out highly correlated features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from scipy import stats

# Load Diabetes dataset
diabetes = load_diabetes(as_frame=True)
df_diabetes = diabetes.frame

print("=" * 60)
print("MILESTONE 2: Data Preparation & Preprocessing")
print("=" * 60)

# Display dataset info
print("\n1. DATASET INFORMATION:")
print(f"Dataset shape: {df_diabetes.shape}")
print(f"Features: {list(df_diabetes.columns)}")

# Check for missing values
print("\n2. MISSING VALUES ANALYSIS:")
missing_values = df_diabetes.isnull().sum()
print("Missing values in each column:")
for col, missing_count in missing_values.items():
    print(f"  {col}: {missing_count}")

# Check for duplicates
print(f"\n3. DUPLICATE RECORDS: {df_diabetes.duplicated().sum()}")

# Handle outliers using IQR method
print("\n4. OUTLIER DETECTION AND HANDLING:")
outliers_info = {}

for column in df_diabetes.columns:
    Q1 = df_diabetes[column].quantile(0.25)
    Q3 = df_diabetes[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_diabetes[(df_diabetes[column] < lower_bound) | (df_diabetes[column] > upper_bound)]
    outlier_count = len(outliers)
    outliers_info[column] = outlier_count

    print(f"  {column}: {outlier_count} outliers ({outlier_count/len(df_diabetes)*100:.2f}%)")

# Apply winsorization to handle outliers (capping at 5th and 95th percentiles)
df_cleaned = df_diabetes.copy()
for column in df_cleaned.columns:
    lower = df_cleaned[column].quantile(0.05)
    upper = df_cleaned[column].quantile(0.95)
    df_cleaned[column] = np.clip(df_cleaned[column], lower, upper)

print("\n5. FEATURE SCALING/NORMALIZATION:")
# Apply StandardScaler to normalize features
scaler = StandardScaler()
features_to_scale = [col for col in df_cleaned.columns if col != 'target']  # Assuming 'target' is the outcome variable

# Scale the features
scaled_features = scaler.fit_transform(df_cleaned[features_to_scale])
df_scaled = pd.DataFrame(scaled_features, columns=features_to_scale)
df_scaled['target'] = df_cleaned['target'].values

print("Features scaled using StandardScaler (mean=0, std=1)")
print("Scaled features statistics:")
print(df_scaled[features_to_scale].describe().round(3))

print("\n6. CATEGORICAL VARIABLES ENCODING:")
# Check for categorical variables
categorical_cols = df_scaled.select_dtypes(include=['object', 'category']).columns
if len(categorical_cols) > 0:
    print(f"Categorical variables found: {list(categorical_cols)}")
    # Apply one-hot encoding if categorical variables exist
    df_encoded = pd.get_dummies(df_scaled, columns=categorical_cols, drop_first=True)
    print("Categorical variables encoded using one-hot encoding")
else:
    df_encoded = df_scaled.copy()
    print("No categorical variables found in the dataset")

print("\n7. CORRELATION ANALYSIS:")
# Calculate correlation matrix
correlation_matrix = df_encoded.corr()

# Display highly correlated features (absolute correlation > 0.7)
print("Highly correlated features (|correlation| > 0.7):")
high_corr_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_value = abs(correlation_matrix.iloc[i, j])
        if corr_value > 0.7:
            feature1 = correlation_matrix.columns[i]
            feature2 = correlation_matrix.columns[j]
            actual_corr = correlation_matrix.iloc[i, j]
            high_corr_pairs.append((feature1, feature2, actual_corr))
            print(f"  {feature1} & {feature2}: {actual_corr:.3f}")

if not high_corr_pairs:
    print("  No highly correlated feature pairs found (|correlation| > 0.7)")

# Visualize correlation matrix
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Feature Correlation Matrix (Diabetes Dataset)')
plt.tight_layout()
plt.show()

# Display top correlations with target variable (assuming 'target' is the outcome)
if 'target' in df_encoded.columns:
    target_correlations = correlation_matrix['target'].drop('target').sort_values(key=abs, ascending=False)
    print(f"\nTop features correlated with target variable:")
    for feature, corr in target_correlations.head(5).items():
        print(f"  {feature}: {corr:.3f}")

print("\n8. FINAL DATASET SUMMARY:")
print(f"Original dataset shape: {df_diabetes.shape}")
print(f"Processed dataset shape: {df_encoded.shape}")
print(f"Features after preprocessing: {list(df_encoded.columns)}")

# Display first few rows of processed data
print("\nFirst 5 rows of processed dataset:")
print(df_encoded.head().round(3))

print("\n" + "=" * 60)
print("DATA PREPARATION COMPLETED SUCCESSFULLY!")
print("=" * 60)
```

Milestone 3: Model Selection & Training
- Split the dataset into training and test sets with a suitable ratio
- Choose and train at least two machine learning models (e.g., logistic regression, random forest)
- Include hyperparameter tuning with cross-validation for one chosen model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("MILESTONE 3: Model Selection & Training")
print("=" * 60)

# Load and prepare the data (using the processed data from Milestone 2)
diabetes = load_diabetes(as_frame=True)
df = diabetes.frame

# For classification, we need to create a binary target variable
# Convert the continuous target to binary (0: no diabetes, 1: diabetes)
# Using median as threshold for binary classification
target_median = df['target'].median()
df['diabetes_class'] = (df['target'] > target_median).astype(int)

print(f"Target variable created: 'diabetes_class'")
print(f"Class distribution:")
print(df['diabetes_class'].value_counts())
print(f"Positive class (diabetes): {(df['diabetes_class'].mean() * 100):.1f}%")

# Prepare features and target
X = df.drop(['target', 'diabetes_class'], axis=1)
y = df['diabetes_class']

print(f"\n1. DATASET SPLITTING:")
print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Training set: {X_train.shape[0]} samples ({X_train.shape[0]/len(X)*100:.1f}%)")
print(f"Test set: {X_test.shape[0]} samples ({X_test.shape[0]/len(X)*100:.1f}%)")
print(f"Training set class distribution:")
print(y_train.value_counts(normalize=True).round(3))

# Feature scaling for models that require it
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n2. MODEL TRAINING:")
print("-" * 40)

# Initialize models
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Support Vector Machine': SVC(random_state=42)
}

# Train models and evaluate with cross-validation
results = {}

print("\nCROSS-VALIDATION RESULTS (5-fold):")
print("-" * 50)

for name, model in models.items():
    if name in ['Logistic Regression', 'Support Vector Machine']:
        # Use scaled data for these models
        X_tr = X_train_scaled
        X_te = X_test_scaled
    else:
        # Use original data for tree-based models
        X_tr = X_train
        X_te = X_test

    # Cross-validation
    cv_scores = cross_val_score(model, X_tr, y_train, cv=5, scoring='accuracy')

    # Train the model
    model.fit(X_tr, y_train)

    # Make predictions
    y_pred = model.predict(X_te)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    results[name] = {
        'model': model,
        'cv_mean': cv_scores.mean(),
        'cv_std': cv_scores.std(),
        'test_accuracy': accuracy,
        'predictions': y_pred
    }

    print(f"\n{name}:")
    print(f"  Cross-validation Accuracy: {cv_scores.mean():.3f} (+/- {cv_scores.std() * 2:.3f})")
    print(f"  Test Accuracy: {accuracy:.3f}")

print("\n3. HYPERPARAMETER TUNING (Random Forest):")
print("-" * 50)

# Hyperparameter tuning for Random Forest using GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

print("Performing grid search with cross-validation...")
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)

print("\nGRID SEARCH RESULTS:")
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_:.3f}")

# Train the tuned model
best_rf = grid_search.best_estimator_
y_pred_tuned = best_rf.predict(X_test)
tuned_accuracy = accuracy_score(y_test, y_pred_tuned)

print(f"Tuned Random Forest Test Accuracy: {tuned_accuracy:.3f}")

# Add tuned model to results
results['Random Forest (Tuned)'] = {
    'model': best_rf,
    'cv_mean': grid_search.best_score_,
    'cv_std': 0,  # Not calculated separately
    'test_accuracy': tuned_accuracy,
    'predictions': y_pred_tuned
}

print("\n4. MODEL COMPARISON:")
print("-" * 40)

# Create comparison DataFrame
comparison_df = pd.DataFrame({
    'Model': list(results.keys()),
    'CV Accuracy': [results[model]['cv_mean'] for model in results.keys()],
    'Test Accuracy': [results[model]['test_accuracy'] for model in results.keys()]
})

comparison_df = comparison_df.sort_values('Test Accuracy', ascending=False)
print(comparison_df.round(3))

# Visualize model performance
plt.figure(figsize=(12, 6))

# Plot 1: Model Accuracy Comparison
plt.subplot(1, 2, 1)
models_list = list(results.keys())
cv_acc = [results[model]['cv_mean'] for model in models_list]
test_acc = [results[model]['test_accuracy'] for model in models_list]

x = np.arange(len(models_list))
width = 0.35

plt.bar(x - width/2, cv_acc, width, label='CV Accuracy', alpha=0.7)
plt.bar(x + width/2, test_acc, width, label='Test Accuracy', alpha=0.7)

plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Performance Comparison')
plt.xticks(x, models_list, rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Feature Importance for Random Forest
plt.subplot(1, 2, 2)
feature_importance = best_rf.feature_importances_
feature_names = X.columns
importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': feature_importance
}).sort_values('importance', ascending=True)

plt.barh(importance_df['feature'], importance_df['importance'])
plt.xlabel('Feature Importance')
plt.title('Random Forest Feature Importance')
plt.tight_layout()

plt.show()

print("\n5. DETAILED PERFORMANCE ANALYSIS:")
print("-" * 50)

# Display detailed classification report for the best model
best_model_name = comparison_df.iloc[0]['Model']
best_model = results[best_model_name]['model']
best_predictions = results[best_model_name]['predictions']

print(f"Best Model: {best_model_name}")
print(f"Test Accuracy: {results[best_model_name]['test_accuracy']:.3f}")
print("\nClassification Report:")
print(classification_report(y_test, best_predictions))

# Confusion Matrix for best model
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, best_predictions)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Diabetes', 'Diabetes'],
            yticklabels=['No Diabetes', 'Diabetes'])
plt.title(f'Confusion Matrix - {best_model_name}')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

print("\n6. PREDICTION EXAMPLES:")
print("-" * 40)

# Show some example predictions
example_indices = np.random.choice(len(X_test), 10, replace=False)
example_predictions = best_predictions[example_indices]
example_actual = y_test.iloc[example_indices].values

example_df = pd.DataFrame({
    'Actual': example_actual,
    'Predicted': example_predictions,
    'Correct': example_actual == example_predictions
})

print(example_df.head(10))
print(f"\nCorrect predictions: {example_df['Correct'].sum()}/10")

print("\n" + "=" * 60)
print("MODEL TRAINING COMPLETED SUCCESSFULLY!")
print("=" * 60)
```

Milestone 4: Model Evaluation & Interpretation
- Evaluate the trained models with appropriate metrics (accuracy, ROC-AUC, precision, recall)
- Plot and interpret a confusion matrix for the best model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                           roc_auc_score, confusion_matrix, classification_report,
                           roc_curve, precision_recall_curve, auc)
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("MILESTONE 4: Model Evaluation & Interpretation")
print("=" * 60)

# Load and prepare the data
diabetes = load_diabetes(as_frame=True)
df = diabetes.frame

# Create binary classification target
target_median = df['target'].median()
df['diabetes_class'] = (df['target'] > target_median).astype(int)

# Prepare features and target
X = df.drop(['target', 'diabetes_class'], axis=1)
y = df['diabetes_class']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("1. MODEL TRAINING FOR EVALUATION:")
print("-" * 40)

# Train multiple models for comprehensive evaluation
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Support Vector Machine': SVC(probability=True, random_state=42)
}

# Store results
evaluation_results = {}

print("Training models for evaluation...")
for name, model in models.items():
    if name in ['Logistic Regression', 'Support Vector Machine']:
        X_tr = X_train_scaled
        X_te = X_test_scaled
    else:
        X_tr = X_train
        X_te = X_test

    # Train model
    model.fit(X_tr, y_train)

    # Get predictions and probabilities
    y_pred = model.predict(X_te)
    y_pred_proba = model.predict_proba(X_te)[:, 1] if hasattr(model, 'predict_proba') else None

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba) if y_pred_proba is not None else None

    evaluation_results[name] = {
        'model': model,
        'predictions': y_pred,
        'probabilities': y_pred_proba,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'roc_auc': roc_auc
    }

    print(f"\n{name}:")
    print(f"  Accuracy:  {accuracy:.3f}")
    print(f"  Precision: {precision:.3f}")
    print(f"  Recall:    {recall:.3f}")
    print(f"  F1-Score:  {f1:.3f}")
    print(f"  ROC-AUC:   {roc_auc:.3f}" if roc_auc else "  ROC-AUC:   Not available")

print("\n2. COMPREHENSIVE METRICS COMPARISON:")
print("-" * 50)

# Create detailed comparison table
metrics_df = pd.DataFrame({
    'Model': list(evaluation_results.keys()),
    'Accuracy': [evaluation_results[model]['accuracy'] for model in evaluation_results.keys()],
    'Precision': [evaluation_results[model]['precision'] for model in evaluation_results.keys()],
    'Recall': [evaluation_results[model]['recall'] for model in evaluation_results.keys()],
    'F1-Score': [evaluation_results[model]['f1_score'] for model in evaluation_results.keys()],
    'ROC-AUC': [evaluation_results[model]['roc_auc'] for model in evaluation_results.keys()]
})

# Sort by F1-Score (balanced metric)
metrics_df = metrics_df.sort_values('F1-Score', ascending=False)
print(metrics_df.round(3))

# Identify best model
best_model_name = metrics_df.iloc[0]['Model']
best_model = evaluation_results[best_model_name]['model']
best_predictions = evaluation_results[best_model_name]['predictions']
best_probabilities = evaluation_results[best_model_name]['probabilities']

print(f"\nBest performing model: {best_model_name}")

print("\n3. DETAILED EVALUATION OF BEST MODEL:")
print("-" * 50)

print(f"\nClassification Report for {best_model_name}:")
print(classification_report(y_test, best_predictions,
                          target_names=['No Diabetes', 'Diabetes']))

# Calculate additional metrics for best model
cm = confusion_matrix(y_test, best_predictions)
tn, fp, fn, tp = cm.ravel()

print(f"\nConfusion Matrix Analysis:")
print(f"True Negatives (TN): {tn} - Correctly predicted no diabetes")
print(f"False Positives (FP): {fp} - Incorrectly predicted diabetes")
print(f"False Negatives (FN): {fn} - Incorrectly predicted no diabetes")
print(f"True Positives (TP): {tp} - Correctly predicted diabetes")

# Calculate derived metrics
specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
false_positive_rate = fp / (fp + tn) if (fp + tn) > 0 else 0
false_negative_rate = fn / (fn + tp) if (fn + tp) > 0 else 0

print(f"\nAdditional Metrics for {best_model_name}:")
print(f"Specificity (True Negative Rate): {specificity:.3f}")
print(f"False Positive Rate: {false_positive_rate:.3f}")
print(f"False Negative Rate: {false_negative_rate:.3f}")
print(f"Prevalence: {y_test.mean():.3f}")

print("\n4. CONFUSION MATRIX VISUALIZATION:")
print("-" * 40)

# Create enhanced confusion matrix plot
plt.figure(figsize=(15, 5))

# Plot 1: Standard Confusion Matrix
plt.subplot(1, 3, 1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted No Diabetes', 'Predicted Diabetes'],
            yticklabels=['Actual No Diabetes', 'Actual Diabetes'],
            cbar_kws={'shrink': 0.8})
plt.title(f'Confusion Matrix - {best_model_name}\n', fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontweight='bold')
plt.xlabel('Predicted Label', fontweight='bold')

# Plot 2: Normalized Confusion Matrix
plt.subplot(1, 3, 2)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_normalized, annot=True, fmt='.2%', cmap='Greens',
            xticklabels=['Predicted No Diabetes', 'Predicted Diabetes'],
            yticklabels=['Actual No Diabetes', 'Actual Diabetes'],
            cbar_kws={'shrink': 0.8})
plt.title(f'Normalized Confusion Matrix - {best_model_name}\n', fontsize=14, fontweight='bold')
plt.ylabel('True Label', fontweight='bold')
plt.xlabel('Predicted Label', fontweight='bold')

# Plot 3: Performance Metrics Breakdown
plt.subplot(1, 3, 3)
metrics_values = [evaluation_results[best_model_name]['accuracy'],
                 evaluation_results[best_model_name]['precision'],
                 evaluation_results[best_model_name]['recall'],
                 evaluation_results[best_model_name]['f1_score'],
                 specificity]
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'Specificity']

colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#3E885B']
bars = plt.bar(metrics_names, metrics_values, color=colors, alpha=0.7)

# Add value labels on bars
for bar, value in zip(bars, metrics_values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{value:.3f}', ha='center', va='bottom', fontweight='bold')

plt.title(f'Performance Metrics - {best_model_name}\n', fontsize=14, fontweight='bold')
plt.ylabel('Score', fontweight='bold')
plt.xticks(rotation=45)
plt.ylim(0, 1.1)
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

print("\n5. ROC-AUC AND PRECISION-RECALL CURVES:")
print("-" * 45)

# Create ROC and Precision-Recall curves
plt.figure(figsize=(15, 6))

# Plot 1: ROC Curves for all models
plt.subplot(1, 2, 1)
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier', alpha=0.6)

for name, results in evaluation_results.items():
    if results['probabilities'] is not None:
        fpr, tpr, _ = roc_curve(y_test, results['probabilities'])
        roc_auc = results['roc_auc']
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.3f})', linewidth=2)

plt.xlabel('False Positive Rate', fontweight='bold')
plt.ylabel('True Positive Rate (Recall)', fontweight='bold')
plt.title('ROC Curves - Model Comparison', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

# Plot 2: Precision-Recall Curves
plt.subplot(1, 2, 2)
for name, results in evaluation_results.items():
    if results['probabilities'] is not None:
        precision_curve, recall_curve, _ = precision_recall_curve(y_test, results['probabilities'])
        pr_auc = auc(recall_curve, precision_curve)
        plt.plot(recall_curve, precision_curve, label=f'{name} (AUC = {pr_auc:.3f})', linewidth=2)

# Add no-skill line
no_skill = len(y_test[y_test==1]) / len(y_test)
plt.plot([0, 1], [no_skill, no_skill], 'k--', label='No Skill', alpha=0.6)

plt.xlabel('Recall', fontweight='bold')
plt.ylabel('Precision', fontweight='bold')
plt.title('Precision-Recall Curves', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

plt.tight_layout()
plt.show()

print("\n6. MODEL INTERPRETATION AND INSIGHTS:")
print("-" * 45)

print(f"\nKey Insights for {best_model_name}:")

# Performance interpretation
accuracy = evaluation_results[best_model_name]['accuracy']
precision = evaluation_results[best_model_name]['precision']
recall = evaluation_results[best_model_name]['recall']
f1 = evaluation_results[best_model_name]['f1_score']

print(f" Overall Accuracy: {accuracy:.1%} - Model correctly classifies {accuracy:.1%} of all cases")
print(f" Precision: {precision:.1%} - When model predicts diabetes, it's correct {precision:.1%} of the time")
print(f" Recall: {recall:.1%} - Model identifies {recall:.1%} of actual diabetes cases")
print(f" F1-Score: {f1:.1%} - Balanced measure of precision and recall")

# Business impact analysis
print(f"\nBusiness Impact Analysis:")
print(f" False Positives ({fp} cases): Non-diabetic patients incorrectly flagged - may cause unnecessary concern")
print(f" False Negatives ({fn} cases): Diabetic patients missed - potentially serious health risk")
print(f" True Positives ({tp} cases): Correctly identified diabetes cases - enables timely intervention")
print(f" True Negatives ({tn} cases): Correctly identified non-diabetic cases - avoids unnecessary treatment")

# Recommendations based on model performance
print(f"\nRecommendations:")
if precision > recall:
    print(" Model is more conservative - better at avoiding false alarms but may miss some cases")
    print(" Consider lowering classification threshold if catching all diabetes cases is critical")
else:
    print(" Model is more sensitive - better at catching true cases but may have more false alarms")
    print(" Consider increasing classification threshold if reducing false positives is important")

print(f"\n7. THRESHOLD ANALYSIS:")
print("-" * 30)

# Threshold analysis for probabilistic models
if best_probabilities is not None:
    thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]
    print(f"\nPerformance at different classification thresholds:")
    print("Threshold | Precision | Recall   | F1-Score")
    print("-" * 40)

    for threshold in thresholds:
        y_pred_thresh = (best_probabilities >= threshold).astype(int)
        precision_t = precision_score(y_test, y_pred_thresh, zero_division=0)
        recall_t = recall_score(y_test, y_pred_thresh, zero_division=0)
        f1_t = f1_score(y_test, y_pred_thresh, zero_division=0)
        print(f"{threshold:9.1f} | {precision_t:9.3f} | {recall_t:8.3f} | {f1_t:8.3f}")

print("\n" + "=" * 60)
print("MODEL EVALUATION COMPLETED SUCCESSFULLY!")
print("=" * 60)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import json
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("MILESTONE 5: Deployment & Reporting")
print("=" * 60)

# Load and prepare the data
diabetes = load_diabetes(as_frame=True)
df = diabetes.frame

# Create binary classification target
target_median = df['target'].median()
df['diabetes_class'] = (df['target'] > target_median).astype(int)

# Prepare features and target
X = df.drop(['target', 'diabetes_class'], axis=1)
y = df['diabetes_class']

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("1. TRAINING FINAL MODEL FOR DEPLOYMENT:")
print("-" * 40)

# Train the best model (Random Forest based on previous milestones)
final_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=42
)

final_model.fit(X_train, y_train)

# Evaluate final model
y_pred = final_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Final Model: Random Forest Classifier")
print(f"Test Accuracy: {accuracy:.3f}")
print(f"Training Samples: {X_train.shape[0]}")
print(f"Feature Count: {X_train.shape[1]}")

print("\n2. SAVING THE TRAINED MODEL:")
print("-" * 35)

# Create feature descriptions manually since DESCRIPTION is not available
feature_descriptions = {
    'age': 'Age in years (standardized)',
    'sex': 'Gender (standardized)',
    'bmi': 'Body Mass Index (standardized)',
    'bp': 'Average Blood Pressure (standardized)',
    's1': 'Total Serum Cholesterol (standardized)',
    's2': 'Low-Density Lipoproteins (LDL) (standardized)',
    's3': 'High-Density Lipoproteins (HDL) (standardized)',
    's4': 'Total Cholesterol / HDL Ratio (standardized)',
    's5': 'Log of Serum Triglycerides Level (standardized)',
    's6': 'Blood Sugar Level (standardized)'
}

# Create a dictionary to store model and metadata
model_artifacts = {
    'model': final_model,
    'feature_names': list(X.columns),
    'feature_descriptions': feature_descriptions,
    'target_median': target_median,
    'model_type': 'RandomForestClassifier',
    'training_date': pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S'),
    'performance': {
        'accuracy': accuracy,
        'feature_count': X_train.shape[1],
        'training_samples': X_train.shape[0]
    }
}

# Save the model using joblib
model_filename = 'diabetes_prediction_model.joblib'
joblib.dump(model_artifacts, model_filename)
print(f" Model saved as: {model_filename}")

# Save feature names separately for reference
feature_info = {
    'feature_names': list(X.columns),
    'feature_descriptions': feature_descriptions,
    'data_shape': {
        'original_samples': len(df),
        'training_samples': len(X_train),
        'test_samples': len(X_test),
        'n_features': X.shape[1]
    }
}

with open('model_features.json', 'w') as f:
    json.dump(feature_info, f, indent=2)

print(" Feature information saved as: model_features.json")

print("\n3. MODEL VERIFICATION - LOADING AND TESTING:")
print("-" * 45)

# Load the saved model
print("Loading saved model...")
loaded_artifacts = joblib.load(model_filename)
loaded_model = loaded_artifacts['model']
loaded_features = loaded_artifacts['feature_names']

print(f" Model loaded successfully")
print(f" Model type: {loaded_artifacts['model_type']}")
print(f" Training date: {loaded_artifacts['training_date']}")
print(f" Features: {len(loaded_features)}")

# Verify the loaded model works
loaded_predictions = loaded_model.predict(X_test)
loaded_accuracy = accuracy_score(y_test, loaded_predictions)

print(f" Loaded model accuracy: {loaded_accuracy:.3f}")
print(f" Original model accuracy: {accuracy:.3f}")
print(f" Model verification: {'SUCCESS' if abs(loaded_accuracy - accuracy) < 0.001 else 'FAILED'}")

print("\n4. MAKING NEW PREDICTIONS:")
print("-" * 30)

# Create new sample data for prediction
print("Creating new sample data for prediction...")

# Method 1: Random samples from test set
sample_indices = np.random.choice(len(X_test), 5, replace=False)
new_samples = X_test.iloc[sample_indices].copy()
actual_labels = y_test.iloc[sample_indices].values

print(f"\nSample predictions from test set:")
print("=" * 60)

predictions = loaded_model.predict(new_samples)
probabilities = loaded_model.predict_proba(new_samples)

for i, (idx, sample) in enumerate(new_samples.iterrows()):
    pred = predictions[i]
    prob = probabilities[i][1]  # Probability of class 1 (diabetes)
    actual = actual_labels[i]

    status = " CORRECT" if pred == actual else " INCORRECT"
    risk_level = "HIGH" if prob > 0.7 else "MEDIUM" if prob > 0.3 else "LOW"

    print(f"\nSample {i+1}:")
    print(f"  Prediction: {'Diabetes' if pred == 1 else 'No Diabetes'}")
    print(f"  Probability: {prob:.3f} ({risk_level} risk)")
    print(f"  Actual: {'Diabetes' if actual == 1 else 'No Diabetes'}")
    print(f"  Status: {status}")

# Method 2: Create synthetic new patient data
print(f"\nSynthetic new patient predictions:")
print("=" * 50)

# Create realistic synthetic patients
synthetic_patients = pd.DataFrame({
    'age': [0.2, -0.1, 0.4, -0.05, 0.3],
    'sex': [0.05, -0.05, 0.1, -0.1, 0.05],
    'bmi': [0.3, -0.2, 0.5, 0.1, 0.4],
    'bp': [0.25, -0.15, 0.35, 0.05, 0.3],
    's1': [0.1, -0.1, 0.2, -0.05, 0.15],
    's2': [0.15, -0.2, 0.25, 0.0, 0.2],
    's3': [-0.1, 0.05, -0.2, 0.1, -0.15],
    's4': [0.2, -0.1, 0.3, 0.05, 0.25],
    's5': [0.25, -0.05, 0.4, 0.1, 0.3],
    's6': [0.15, -0.15, 0.25, 0.0, 0.2]
})

# Ensure columns match training data
synthetic_patients = synthetic_patients[loaded_features]

# Make predictions
synth_predictions = loaded_model.predict(synthetic_patients)
synth_probabilities = loaded_model.predict_proba(synthetic_patients)

for i in range(len(synthetic_patients)):
    pred = synth_predictions[i]
    prob = synth_probabilities[i][1]
    risk_level = "HIGH" if prob > 0.7 else "MEDIUM" if prob > 0.3 else "LOW"

    print(f"\nNew Patient {i+1}:")
    print(f"  Prediction: {'DIABETES' if pred == 1 else 'NO DIABETES'}")
    print(f"  Confidence: {prob:.1%}")
    print(f"  Risk Level: {risk_level}")

    # Provide interpretation
    if pred == 1:
        print(f"  Recommendation: Consult healthcare provider for further testing")
    else:
        print(f"  Recommendation: Maintain healthy lifestyle with regular checkups")

print("\n5. DEPLOYMENT ARTIFACTS SUMMARY:")
print("-" * 35)

# List all saved files and their purposes
deployment_files = {
    'diabetes_prediction_model.joblib': 'Main trained model with metadata',
    'model_features.json': 'Feature information and descriptions',
}

print("Deployment files created:")
for file, purpose in deployment_files.items():
    print(f"   {file}: {purpose}")

print("\n6. PREDICTION API SIMULATION:")
print("-" * 30)

# Simulate a simple prediction function for deployment
def predict_diabetes_risk(patient_data, model_artifacts):
    """
    Simulated API function for diabetes prediction

    Args:
        patient_data: Dictionary or DataFrame with patient features
        model_artifacts: Loaded model artifacts

    Returns:
        Dictionary with prediction and metadata
    """
    model = model_artifacts['model']
    feature_names = model_artifacts['feature_names']

    # Convert to DataFrame and ensure correct feature order
    if isinstance(patient_data, dict):
        patient_df = pd.DataFrame([patient_data])[feature_names]
    else:
        patient_df = patient_data[feature_names]

    prediction = model.predict(patient_df)[0]
    probability = model.predict_proba(patient_df)[0][1]

    return {
        'prediction': int(prediction),
        'probability': float(probability),
        'risk_level': 'HIGH' if probability > 0.7 else 'MEDIUM' if probability > 0.3 else 'LOW',
        'interpretation': 'Diabetes risk detected' if prediction == 1 else 'No significant diabetes risk',
        'timestamp': pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')
    }

# Test the prediction function
print("Testing prediction function...")
test_patient = {
    'age': 0.25,
    'sex': 0.05,
    'bmi': 0.35,
    'bp': 0.28,
    's1': 0.12,
    's2': 0.18,
    's3': -0.12,
    's4': 0.22,
    's5': 0.28,
    's6': 0.18
}

result = predict_diabetes_risk(test_patient, loaded_artifacts)
print(f"\nPrediction API Result:")
for key, value in result.items():
    print(f"  {key}: {value}")

print("\n7. MODEL PERFORMANCE REPORT:")
print("-" * 30)

# Generate final performance report
feature_importance = loaded_model.feature_importances_
feature_importance_df = pd.DataFrame({
    'feature': loaded_features,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

print("\nTop 5 Most Important Features:")
for i, row in feature_importance_df.head().iterrows():
    print(f"  {row['feature']}: {row['importance']:.3f}")

# Final classification report
print(f"\nFinal Model Performance Summary:")
print(f"Accuracy: {accuracy:.3f}")
print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
print(f"Feature count: {X_train.shape[1]}")

print(f"\nClass Distribution:")
print(f"Diabetes cases: {y_test.sum()} ({y_test.mean():.1%})")
print(f"Non-diabetes cases: {len(y_test) - y_test.sum()} ({1 - y_test.mean():.1%})")

print("\n8. USAGE EXAMPLE FOR FUTURE REFERENCE:")
print("-" * 45)

print("""
# How to use the saved model in future:

import joblib
import pandas as pd

# Load the model
model_artifacts = joblib.load('diabetes_prediction_model.joblib')
model = model_artifacts['model']
feature_names = model_artifacts['feature_names']

# Prepare new patient data
new_patient = {
    'age': 0.1, 'sex': -0.05, 'bmi': 0.2, 'bp': 0.15,
    's1': 0.08, 's2': 0.12, 's3': -0.08, 's4': 0.18,
    's5': 0.22, 's6': 0.1
}

# Make prediction
new_patient_df = pd.DataFrame([new_patient])[feature_names]
prediction = model.predict(new_patient_df)[0]
probability = model.predict_proba(new_patient_df)[0][1]

print(f"Prediction: {'Diabetes' if prediction == 1 else 'No Diabetes'}")
print(f"Probability: {probability:.3f}")
""")

print("\n" + "=" * 60)
print("DEPLOYMENT COMPLETED SUCCESSFULLY!")
print("=" * 60)
print("\nNEXT STEPS:")
print(" Model saved as 'diabetes_prediction_model.joblib'")
print(" Use joblib.load() to load the model in production")
print(" Ensure input data has same feature structure")
print(" Monitor model performance over time")
print(" Consider retraining with new data periodically")
```
